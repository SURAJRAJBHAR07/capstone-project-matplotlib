```python
import pandas as pd
import matplotlib.pyplot as plt

#load the data
df=pd.read_csv('netflix_retail_customers.csv')
print(df)
```

```python
#clean data
df=df.dropna(subset=['Customer_ID','Age','Gender','Country','Subscription_Type','Monthly_Fee','Watch_Hours_Per_Week','Preferred_Genre','Churned'])

type_counts=df['Preferred_Genre'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_counts.index, type_counts.values , color=['skyblue','orange'])
plt.title('NETFLIX PROJECT')
plt.xlabel('Type Of Shows')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('CAPSTONE PROJECT IN MATPLOTLIB')
plt.show()
             
```

```python
#clean data
df=df.dropna(subset=['Customer_ID','Age','Gender','Country','Subscription_Type','Monthly_Fee','Watch_Hours_Per_Week','Preferred_Genre','Churned'])

country_counts=df['Country'].value_counts()
plt.figure(figsize=(6,4))
plt.pie(country_counts,labels=country_counts.index, colors=['skyblue','orange','green','red','purple'],autopct='%1.1f%%')
plt.title('Countries Data')
plt.xlabel('Countries')
plt.ylabel('Count Of People')
plt.tight_layout()
plt.savefig('CAPSTONE PROJECT IN MATPLOTLIB')
plt.show()
```

```python
movie_df=df[df['Customer_ID']=='Id'].copy()
movie_df['Watch_Hours_Per_Week']=movie_df['Watch_Hours_Per_Week']

plt.figure(figsize=(8,6))
plt.hist(df[ 'Watch_Hours_Per_Week'],bins=30,color='purple',edgecolor='black')
plt.title('Duration Of movie Duration')
plt.xlabel('Duration Minutes')
plt.ylabel('Number of movies')
plt.tight_layout()
plt.savefig('Movie duration histogram.png')
plt.show()
```

```python
subscription_counts=df['Subscription_Type'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.bar(subscription_counts.index, subscription_counts.values,color='orange')
plt.title('SUBCRIPTION TYPE')
plt.xlabel('SUBSCRIPTION TYPE')
plt.ylabel('Number of movies')
plt.tight_layout()
plt.savefig('SUBSCRIPTION_TYPE_SCATTER.png')
plt.show()
```

```python

country_counts=df['Country'].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(country_counts.index,country_counts.values, color='teal')
plt.title('Countries Data')
plt.xlabel('Countries')
plt.ylabel('Count Of People')
plt.tight_layout()
plt.savefig('CAPSTONE PROJECT IN MATPLOTLIB')
plt.show()
content_by_year=df.groupby(['Watch_Hours_Per_Week','Preferred_Genre']).size().unstack().fillna(0)

fig, ax=plt.subplots(1,2, figsize=(12,5))

#first subplot.movies
ax[0].plot(content_by_year.index,content_by_year.sum(axis=1),color='blue')
ax[0].set_title('Movies Release Per Year')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Number of Movies')
plt.tight_layout()
plt.savefig('movies_tv_shows')
plt.show()

fig, ax = plt.subplots(1, 2, figsize=(12,5))

# First subplot (all genres)
content_by_year.plot(ax=ax[0])
ax[0].set_title('Genre Trends')

# Second subplot (example: Drama only)
ax[1].plot(content_by_year.index, content_by_year['Drama'], color='orange')
ax[1].set_title('Drama Trend')

plt.tight_layout()
plt.show()
```
